```python
# prompt: hazme un programa para descomponer un numero entero en productos de potencias de nmeros primos en espaol

def factorizar_primos(n):
  """
  Descompone un nmero entero en producto de potencias de nmeros primos.

  Args:
    n: El nmero entero a factorizar.

  Returns:
    Un diccionario donde las claves son los factores primos y los valores son
    sus exponentes correspondientes.  Retorna un mensaje de error si la entrada no es vlida.
  """

  if not isinstance(n, int) or n <= 0:
      return "Entrada invlida. Debe ser un entero positivo."

  factores = {}
  d = 2
  while d * d <= n:
    while n % d == 0:
      factores[d] = factores.get(d, 0) + 1
      n //= d
    d += 1
  if n > 1:
    factores[n] = factores.get(n, 0) + 1

  return factores

# Ejemplo de uso
numero = int(input("Ingrese un nmero entero positivo: "))
resultado = factorizar_primos(numero)

if isinstance(resultado, str):  # Verifica si se retorn un mensaje de error
    print(resultado)
else:
    print(f"La factorizacin prima de {numero} es:")
    for factor, exponente in resultado.items():
        print(f"{factor}^{exponente}", end=" ")
    print()
```

$\bullet$ **Paso_9. Todo el algoritmo en un solo lugar**

```python
import matplotlib.pyplot as plt
import pandas as pd
import string


# Convertir el prrafo a minsculas para evitar distincin entre maysculas y minsculas
parrafo = parrafo.lower()

# Crear una lista de letras en el alfabeto
alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

# Inicializar un diccionario para almacenar la frecuencia de cada letra
frecuencias = {letra: 0 for letra in alfabeto}

# Contar la frecuencia de cada letra en el prrafo
for letra in parrafo:
    if letra in alfabeto:
        frecuencias[letra] += 1

# Crear un DataFrame a partir del diccionario de frecuencias
df = pd.DataFrame(frecuencias.values(), index=frecuencias.keys(), columns=['Frecuencia'])

# Crear un grfico de barras
plt.figure(figsize=(10, 4))
df.plot(kind='bar', legend=False)
plt.xlabel('Letra')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de letras en el prrafo')
plt.xticks(rotation=0)
plt.show()

# Crear un grfico de pastel
plt.figure(figsize=(6, 6))
df.plot(kind='pie', y='Frecuencia', autopct='%1.1f%%', legend=False)
plt.title('Frecuencia de letras en el prrafo')
plt.show()
```

$\bullet$ **Paso_10. Todo el algoritmo en un solo lugar pero con las letras que aparecen en el prrafo**

```python
import matplotlib.pyplot as plt
import pandas as pd
import string

print("Letra   Frecuencia")
print("-----------------")
# Prrafo
parrafo = 'La ciencia de datos es un campo acadmico interdisciplinario que utiliza estadstica, computacin cientfica, mtodos, procesos, algoritmos y sistemas cientficos para obtener (recolectar o extraer), tratar, analizar y presentar informes a partir de datos ruidosos, estructurados y no estructurados.'

# Convertir el prrafo a minsculas para contar letras sin distincin entre maysculas y minsculas
parrafo = parrafo.lower()

# Crear un diccionario para almacenar la frecuencia de cada letra
frecuencia_letras = {}

# Iterar a travs del prrafo y contar las letras
for letra in parrafo:
    if letra.isalpha():  # Verificar si es una letra
        if letra in frecuencia_letras:
            frecuencia_letras[letra] += 1
        else:
            frecuencia_letras[letra] = 1

# Mostrar la tabla de frecuencia de letras

for letra, frecuencia in frecuencia_letras.items():
    print(f"{letra}      {frecuencia}")


# Crear un DataFrame a partir del diccionario de frecuencias
df = pd.DataFrame(frecuencia_letras.values(), index=frecuencia_letras.keys(), columns=['Frecuencia'])

# Crear un grfico de barras
plt.figure(figsize=(10, 4))
df.plot(kind='bar', legend=False)
plt.xlabel('Letra')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de letras en el prrafo')
plt.xticks(rotation=0)
plt.show()

# Crear un grfico de pastel
plt.figure(figsize=(6, 6))
df.plot(kind='pie', y='Frecuencia', autopct='%1.1f%%', legend=False)
plt.title('Frecuencia de letras en el prrafo')
plt.show()
```
