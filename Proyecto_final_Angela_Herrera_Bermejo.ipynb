**1. ¿Cuál es la correlación entre edad y definitiva?**

Se puede realizar un gráfico de dispersión y calcular el coeficiente de correlación

```python
# @title **Gráfico de dispersión para observar la relación entre Edad y nota definitiva**

# Gráfico de dispersión para observar la relación entre Edad y nota definitiva
plt.figure(figsize=(8, 6))
plt.scatter(df['EDAD'], df['DEFINITIVA'], color='blue')
plt.title('Relación entre EDAD y Nota Definitiva 2do periodo')
plt.xlabel('EDAD ')
plt.ylabel('DEFINITIVA 2do Periodo')
plt.grid(True)
plt.show()

# Calcular la correlación entre EDAD y nota definitiva
correlacion_edad_nota = df['EDAD'].corr(df['DEFINITIVA'])
print(f'coef. de correlación de Pearson es r = {correlacion_edad_nota:.2f}')
print(f'coef. de determinación R^2 = {(correlacion_edad_nota**2*100):.2f}%')
```

**Analisis**

El gráfico de dispersión muestra la relación entre la edad y la nota definitiva de 2do periodo, y el coeficiente de correlación entre ambas variables es **-0.04**. Este valor sugiere una correlación negativa, es decir se asocian en sentido inverso

```python
# @title **Agregame la recta estimada de regresión y su ecuacion el titulo y el R2 en %**

# ... (Your existing code) ...

# Gráfico de dispersión para observar la relación entre edad y nota definitiva
plt.figure(figsize=(8, 6))
sns.regplot(x='EDAD', y='DEFINITIVA', data=df)  # Use sns.regplot for regression line
plt.title('Relación entre Edad y Nota definitiva 2do Periodo')
plt.xlabel('Edad (año)')
plt.ylabel('Definitiva 2do Periodo')
plt.grid(True)

# Calcular la correlación entre edad y nota definitiva
correlacion_edad_nota = df['EDAD'].corr(df['DEFINITIVA'])

# Obtener la ecuación de la recta de regresión
slope, intercept = np.polyfit(df['DEFINITIVA'], df['EDAD'], 1)
equation = f'y = {slope:.2f}x + {intercept:.2f}'

# Calcular R^2 en porcentaje
r_squared = correlacion_edad_nota ** 2 * 100

# Ajustar las coordenadas para colocar el texto dentro del gráfico
x_coord = df['EDAD'].mean()  # Use the mean of 'EDAD' for x-coordinate
y_coord = df['DEFINITIVA'].min() + (df['DEFINITIVA'].max() - df['DEFINITIVA'].min()) * 0.1  # Adjust y-coordinate

# Agregar la ecuación y R^2 al gráfico en las nuevas coordenadas
plt.text(x_coord, y_coord, equation, fontsize=12)
plt.text(x_coord, y_coord - (df['DEFINITIVA'].max() - df['DEFINITIVA'].min()) * 0.05, f'R^2 = {r_squared:.2%}', fontsize=12)

plt.show()

# ... (Rest of your code) ...
```

**2. ¿Cuál es la correlación entre asistencia y definitiva?**

Realizo un grafico de dispersión y calculo el coeficiente de correlación.

```python
# @title **Agregame la recta estimada de regresión y su ecuacion el titulo y el R2 en %**

# ... (Your existing code) ...

# Gráfico de dispersión para observar la relación entre asistencia y nota definitiva del 2do periodo
plt.figure(figsize=(8, 6))
sns.regplot(x='ASISTENCIA_2DO', y='DEFINITIVA', data=df)  # Use sns.regplot for regression line
plt.title('Relación entre ASISTENCIA y DEFINITIVA en el 2do Periodo')
plt.xlabel('ASISTENCIA_2DO)')
plt.ylabel('DEFINITIVA')
plt.grid(True)

# Calcular la correlación entre asistencia y nota definitiva del 2do periodo
correlacion_ASISTENCIA2DO_nota = df['DEFINITIVA'].corr(df['ASISTENCIA_2DO'])

# Obtener la ecuación de la recta de regresión
slope, intercept = np.polyfit(df['ASISTENCIA_2DO'], df['DEFINITIVA'], 1)
equation = f'y = {slope:.2f}x + {intercept:.2f}'

# Agregar la ecuación de la recta de regresión al gráfico
plt.text(df['ASISTENCIA_2DO'].min(), df['DEFINITIVA'].max() * 0.9, equation, fontsize=12)

# Calcular R^2 en porcentaje
r_squared = correlacion_ASISTENCIA2DO_nota ** 2 * 100

# Agregar R^2 al gráfico
plt.text(df['ASISTENCIA_2DO'].min(), df['DEFINITIVA'].max() * 0.85, f'R² = {r_squared:.2f}%', fontsize=12)


plt.show()

# ... (Rest of your code) ...
```

**8. ¿Qué Area tiene la menor dispersión de notas?**

```python
# Calcular la desviación estándar de las notas por area para medir la dispersión
dispersion_nota_area = df.groupby('AREA')['DEFINITIVA'].std()

# Identificar el area con menor dispersión
area_menor_dispersión = dispersion_nota_area.idxmin()
menor_dispersión_valor = dispersion_nota_area.min()

print(f'El area con la menor dispersión de notas es {area_menor_dispersión} con una desviación estándar de {menor_dispersión_valor:.2f}')
```

El área con la menor dispersión de notas es **C Naturales**, con una desviación estándar de **0.73**.
