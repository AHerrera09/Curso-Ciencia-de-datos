# **Relaciones entre variables**

**1. Cul es la correlacin entre edad y definitiva?**

Se puede realizar un grfico de dispersin y calcular el coeficiente de correlacin

```python
# @title **Grfico de dispersin para observar la relacin entre Edad y nota definitiva**

# Grfico de dispersin para observar la relacin entre Edad y nota definitiva
plt.figure(figsize=(8, 6))
plt.scatter(df['EDAD'], df['DEFINITIVA'], color='blue')
plt.title('Relacin entre EDAD y Nota Definitiva 2do periodo')
plt.xlabel('EDAD ')
plt.ylabel('DEFINITIVA 2do Periodo')
plt.grid(True)
plt.show()

# Calcular la correlacin entre EDAD y nota definitiva
correlacion_edad_nota = df['EDAD'].corr(df['DEFINITIVA'])
print(f'coef. de correlacin de Pearson es r = {correlacion_edad_nota:.2f}')
print(f'coef. de determinacin R^2 = {(correlacion_edad_nota**2*100):.2f}%')
```

**Analisis**

El grfico de dispersin muestra la relacin entre la edad y la nota definitiva de 2do periodo, y el coeficiente de correlacin entre ambas variables es **-0.04**. Este valor sugiere una correlacin negativa, es decir se asocian en sentido inverso

```python
# @title **Agregame la recta estimada de regresin y su ecuacion el titulo y el R2 en %**

# ... (Your existing code) ...

# Grfico de dispersin para observar la relacin entre edad y nota definitiva
plt.figure(figsize=(8, 6))
sns.regplot(x='EDAD', y='DEFINITIVA', data=df)  # Use sns.regplot for regression line
plt.title('Relacin entre Edad y Nota definitiva 2do Periodo')
plt.xlabel('Edad (ao)')
plt.ylabel('Definitiva 2do Periodo')
plt.grid(True)

# Calcular la correlacin entre edad y nota definitiva
correlacion_edad_nota = df['EDAD'].corr(df['DEFINITIVA'])

# Obtener la ecuacin de la recta de regresin
slope, intercept = np.polyfit(df['DEFINITIVA'], df['EDAD'], 1)
equation = f'y = {slope:.2f}x + {intercept:.2f}'

# Calcular R^2 en porcentaje
r_squared = correlacion_edad_nota ** 2 * 100

# Ajustar las coordenadas para colocar el texto dentro del grfico
x_coord = df['EDAD'].mean()  # Use the mean of 'EDAD' for x-coordinate
y_coord = df['DEFINITIVA'].min() + (df['DEFINITIVA'].max() - df['DEFINITIVA'].min()) * 0.1  # Adjust y-coordinate

# Agregar la ecuacin y R^2 al grfico en las nuevas coordenadas
plt.text(x_coord, y_coord, equation, fontsize=12)
plt.text(x_coord, y_coord - (df['DEFINITIVA'].max() - df['DEFINITIVA'].min()) * 0.05, f'R^2 = {r_squared:.2%}', fontsize=12)

plt.show()

# ... (Rest of your code) ...
```

**2. Cul es la correlacin entre asistencia y definitiva?**

Realizo un grafico de dispersin y calculo el coeficiente de correlacin.

```python
# @title **Agregame la recta estimada de regresin y su ecuacion el titulo y el R2 en %**

# ... (Your existing code) ...

# Grfico de dispersin para observar la relacin entre asistencia y nota definitiva del 2do periodo
plt.figure(figsize=(8, 6))
sns.regplot(x='ASISTENCIA_2DO', y='DEFINITIVA', data=df)  # Use sns.regplot for regression line
plt.title('Relacin entre ASISTENCIA y DEFINITIVA en el 2do Periodo')
plt.xlabel('ASISTENCIA_2DO)')
plt.ylabel('DEFINITIVA')
plt.grid(True)

# Calcular la correlacin entre asistencia y nota definitiva del 2do periodo
correlacion_ASISTENCIA2DO_nota = df['DEFINITIVA'].corr(df['ASISTENCIA_2DO'])

# Obtener la ecuacin de la recta de regresin
slope, intercept = np.polyfit(df['ASISTENCIA_2DO'], df['DEFINITIVA'], 1)
equation = f'y = {slope:.2f}x + {intercept:.2f}'

# Agregar la ecuacin de la recta de regresin al grfico
plt.text(df['ASISTENCIA_2DO'].min(), df['DEFINITIVA'].max() * 0.9, equation, fontsize=12)

# Calcular R^2 en porcentaje
r_squared = correlacion_ASISTENCIA2DO_nota ** 2 * 100

# Agregar R^2 al grfico
plt.text(df['ASISTENCIA_2DO'].min(), df['DEFINITIVA'].max() * 0.85, f'R = {r_squared:.2f}%', fontsize=12)


plt.show()

# ... (Rest of your code) ...
```
